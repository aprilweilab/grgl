# Simulating Phenotypes

[grg_pheno_sim](https://github.com/aprilweilab/grg_pheno_sim) is a package that lets you simulate phenotypes on a dataset, with a ton of different options. For more detailed examples, see the [grg_pheno_sim demo notebooks](https://github.com/aprilweilab/grg_pheno_sim/tree/main/demos). Here we just show a few simple scenarios.

`grg_pheno_sim` randomly chooses causal mutations from the dataset (GRG), and gives each chosen mutation a random effect size: how much it influences the final phenotype value $Y$. Then the phenotype itself is calculated from these effect sizes by summing the effects of all the mutations that each individual has. These simulations are useful for testing assumptions about phenotypes in populations, and validating the functionality of any method that is trying to reverse this process (find the causes of phenotypes).


**What you'll need:**

* Python dependencies "grapp", "igdtools", "seaborn", "matplotlib", "numpy": `pip install grapp igdtools seaborn matplotlib numpy`
* Command line tool "wget": `sudo apt install wget` (or your system's equivalent)

### Get Dataset

We'll use the same dataset as for the GWAS tutorial.

```python
%%bash

if [[ ! -e gwas.example.igd ]]; then
  # Download a small example dataset
  wget https://github.com/aprilweilab/grg_pheno_sim/raw/refs/heads/main/demos/data/test-200-samples.vcf.gz -O gwas.example.vcf.gz

  # Convert to IGD; this isn't necessary, but most of the time you will want to do this
  igdtools gwas.example.vcf.gz -o gwas.example.igd
fi

# Just show some stats about the dataset
igdtools -s gwas.example.igd
```

```python
%%bash

if [[ ! -e gwas.example.grg ]]; then
  # -j controls how many threads to use.
  grg construct -j 1 gwas.example.igd -o gwas.example.grg
fi
```
